```
import os
```

```
dir_names = ('train', 'valid', 'test')
category_names = ('goku', 'luffy', 'saitama')
```

```
os.chdir('/content/drive/My Drive/open_ai/anime-classification/data')
```

```
for dir in dir_names:
  for category in category_names:
    src_dir = os.path.join(dir, category)
    orig_filenames = os.listdir(src_dir)
    num_files = len(orig_filenames)
    for filenum, filename in enumerate(orig_filenames):
      src_filepath = os.path.join(src_dir, filename)
      dest_filepath = os.path.join(src_dir, str(filenum) + '.png')
      print('src:', src_filepath)
      print('dest:', dest_filepath)
      os.rename(src_filepath, dest_filepath)
```
